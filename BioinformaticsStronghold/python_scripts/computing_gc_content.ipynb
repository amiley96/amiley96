```python
import re
```

```python
filename = "/Users/alexandriamiley/Desktop/Rosalind/Bioinformatics Stronghold/rosalind_gc.txt"

with open(filename) as f:
    contents = f.readlines()
```

```python
new = '' .join(' '.join(contents).split()).strip()
```

```python
splitted = ''.join(re.split('\\\\',new))
justRos=re.sub('^.*?>Rosalind', '>Rosalind', splitted)
```

```python
justRosList=justRos.split('>')
```

```python
keylist = []
valuelist = []


for element in justRosList:
    if len(element) == 0:
            print(element)
    else:
        keys = str((re.findall('Rosalind_[0-9]*', element))).strip("'[,]'")
        keylist.append(str(keys))
        value = re.split('Rosalind_[0-9]*',element)[1].strip('}')
        valuelist.append(str(value))
```

```python
d = {}
count = 0

for i in keylist:
    if i not in d:
        d[i] = valuelist[count]
        count+=1
        
```

```python
percentlist = []
newdict = {}
for key in d:
    print(d[key])
    newdict[key]=((len(re.findall('G',d[key]))+len(re.findall('C',d[key])))/len(d[key]))

print(newdict)
```

```python
import operator
sorted_x = sorted(newdict.items(), key=operator.itemgetter(0))

print((sorted_x)[0][0])
print(round((sorted_x)[0][1]*100,7))
print(round((sorted_x)[0][1]*100,3))
```
