```python
filename = "/Users/alexandriamiley/Desktop/Rosalind/BioinformaticsStronghold/data/rosalind_prot.txt"

with open(filename) as f:
    contents = f.readlines()
```

```python
rna=contents[0]
```

```python
protdict = {"UUU":"F","CUU" :"L","AUU" :"I","GUU" :"V","UUC" :"F","CUC" :"L","AUC" :"I","GUC":"V","UUA" :"L",
            "CUA" :"L","AUA" :"I","GUA" :"V","UUG" :"L","CUG" :"L","AUG" :"M","GUG" :"V","UCU" :"S","CCU" :"P",
            "ACU" :"T","GCU": "A","UCC" :"S","CCC" :"P","ACC" :"T","GCC": "A","UCA" :"S","CCA" :"P","ACA" :"T",
            "GCA": "A","UCG" :"S","CCG" :"P","ACG" :"T","GCG": "A","UAU" :"Y","CAU" :"H","AAU" :"N","GAU": "D",
            "UAC" :"Y","CAC" :"H","AAC" :"N","GAC": "D","UAA" :"Stop","CAA" :"Q","AAA" :"K","GAA": "E","UAG" :"Stop",
            "CAG" :"Q","AAG" :"K","GAG": "E","UGU" :"C","CGU" :"R","AGU" :"S","GGU": "G","UGC" :"C","CGC" :"R",
            "AGC" :"S","GGC": "G","UGA" :"Stop","CGA" :"R","AGA" :"R","GGA": "G","UGG" :"W","CGG" :"R","AGG" :"R",
            "GGG": "G"}
```

```python
start = 0
stop = 3

protlist = []
protein = ''
```

```python
for codon in range(len(rna)):
    if stop<=len(rna):
        setinquestion=rna[start:stop]
#         print(setinquestion,start,stop)
        protein=(protdict[setinquestion])
        while protein!="Stop":
            try:
                protlist.append(protein)
                break
            except ValueError:
                pass
        start+=3
        stop+=3
```

```python
encodedprotein=''.join(protlist)
print(encodedprotein)
```
